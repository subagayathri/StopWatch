```python
import time
print("press ENTER key to start")
print("press CTRL c to exit")
while True:
   try:
     input()
     print("Started")
     start=time.time()
   except KeyboardInterrupt:
     print("Stopped")
     end=time.time()
     print("Total Time:" ,round(end-start,2), "Seconds")
     break

    
```
